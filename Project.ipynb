# <span style='color:purple'> Fitness Dataset Analysis </span>

##  <span style='color:darkviolet'> Data Discription </span> 

- The given dataset contains information about individuals ' fitness goals and their barriers to achieving those goals'.
- The dataset provide information of every individual past and future expectation from their excercise routine.

## <span style='color:darkviolet'> Data Insights </span>

- **Timestamp:**   The date and time when the survey was filled out.
- **Name:**        The name of the individual who filled out the survey.
- **Gender:**      The gender of the individual.
- **Age:** The age group of the individual.
- **Exercise_Importance:** The level of importance the individual places on exercise.
- **Fitness_Level:** The individual's self-assessment of their current fitness level.
- **Exercise_Frequency:** The frequency of exercise the individual engages in.
- **Barriers_to_Exercise:** The barriers the individual faces in exercising more regularly.
- **Forms_of_Exercise:** The forms of exercise the individual currently participates in.
- **Exercise_Location:** The location where the individual exercises.
- **Exercise_Time:** The preferred time of day for exercise.
- **Exercise_Duration:** The duration of exercise per day.
- **Healthy_Diet:** Whether the individual eats a healthy balanced diet.
- **Diet_Barriers:** The factors that prevent the individual from eating a healthy balanced diet.
- **SelfHealth_Level:** The individual's self-assessment of their overall health.
- **Recommended_to_Friend:** Whether the individual has recommended a fitness routine to friends.
- **Fitness_Equipment:** Whether the individual has purchased fitness equipment.
- **Motivations_to_Exercise:** The motivations for the individual to exercise.

## <span style='color: darkolivegreen'> Bivariate </span>

### <span style='color: indigo'> Exercise Importance According to Self Health Level </span>

### <span style='color: indigo'> Exercise Frequency by Age Group </span>

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Age', y='Exercise_Importance', data=df1,palette='viridis')
plt.title('Exercise Frequency by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

### <span style='color: indigo'> Calculated the count of barriers by age group using Groupby. </span>

```python
b1= df1.groupby('Age')['What barriers, if any, prevent you from exercising more regularly?           (Please select all that apply)'].value_counts().unstack()
b1.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='coolwarm')
plt.title('Barriers to Exercise by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Barriers', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### <span style='color:red'> Observation </span>

- **Different** age groups have **different** barriers to exercising **more** regularly.
- Understanding these barriers can help in designing targeted interventions.
- This stacked bar shows the **Highest** to **Lowest** barriers.

### <span style='color: indigo'> Motivation to Exercise by Age Group and Gender </span>

```python
g2= df1.groupby(['Age', 'Gender'])['Exercise_Location'].value_counts(normalize=True).unstack()
g2.plot(kind='bar', figsize=(12, 10), colormap='Paired')
plt.title('Motivation to Exercise by Age Group and Gender')
plt.xlabel('Age Group and Gender')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Motivation', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
